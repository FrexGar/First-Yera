```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
# Definir categorías
categories = ['Food', 'Travel', 'Fashion', 'Fitness', 'Music', 'Culture', 'Family', 'Health']

# Número de entradas
n = 500

# Crear datos aleatorios
data = {
    'Date': pd.date_range('2021-01-01', periods=n),
    'Category': [random.choice(categories) for _ in range(n)],
    'Likes': np.random.randint(0, 10000, size=n)
}
```

```python
# Cargar datos en un DataFrame
df = pd.DataFrame(data)

# Imprimir la cabeza del DataFrame, información y descripción
print(df.head())
print(df.info())
print(df.describe())

# Contar cada elemento de 'Category'
print(df['Category'].value_counts())
```

```python
# Eliminar valores nulos
df.dropna(inplace=True)

# Eliminar duplicados
df.drop_duplicates(inplace=True)

# Convertir 'Date' a formato datetime
df['Date'] = pd.to_datetime(df['Date'])

# Asegurarnos de que 'Likes' sea un entero
df['Likes'] = df['Likes'].astype(int)
```

```python
# Histograma de Likes
plt.figure(figsize=(10, 6))
sns.histplot(df['Likes'], bins=30)
plt.title('Distribución de Likes')
plt.xlabel('Likes')
plt.ylabel('Frecuencia')
plt.show()

# Boxplot de Likes por categoría
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Likes', data=df)
plt.title('Boxplot de Likes por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Likes')
plt.show()

# Calcular la media de Likes
print("Media de Likes:", df['Likes'].mean())

# Media de Likes por categoría
mean_likes_by_category = df.groupby('Category')['Likes'].mean()
print("Media de Likes por Categoría:\n", mean_likes_by_category)
```
